```python
import pandas as pd
import numpy as np
import rpy2.robjects as robjects
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri

pandas2ri.activate()
```

```python
# Defining the R script and loading the instance in Python
r = robjects.r
r['source']('exec_problem.R')

# Loading the function we have defined in R.
exec_function_r = robjects.globalenv['hydro_prob']
```

```python
result_files = [
    "config0_MSE.csv",
    "config1_MSE.csv",
    "config2_MSE.csv",
    "config3_MSE.csv",
    "config0_NSE.csv",
    "config1_NSE.csv",
    "config2_NSE.csv",
    "config3_NSE.csv",
    "config0_R2.csv",
    "config1_R2.csv",
    "config2_R2.csv",
    "config3_R2.csv",
    "config0_KGE.csv",
    "config1_KGE.csv",
    "config2_KGE.csv",
    "config3_KGE.csv",
]

import os

df = pd.DataFrame(columns=["model", "target", "MSE", "RMSE", "NSE", "R2", "KGE"])
for file_name in result_files:
    print(file_name, "Yes" if os.path.exists(file_name) else "No")
    params = np.loadtxt(file_name, delimiter=",", max_rows=1)
    model = int(file_name[6])
    target = file_name.split(".")[0][8:]
    metrics = exec_function_r(model, params) 
    df.loc[len(df)] = [model, target, metrics[0][0], metrics[1][0], metrics[3][0], metrics[4][0], metrics[5][0]]
    
```
